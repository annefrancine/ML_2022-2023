## Encodeur d'étiquette (label encoder):
L'encodage d'étiquettes en Python peut être réalisé à l'aide de la bibliothèque Sklearn. Sklearn fournit un outil très efficace pour coder les niveaux des caractéristiques catégorielles en valeurs numériques. LabelEncoderencoder des étiquettes avec une valeur comprise entre 0 et n_classes-1 où n est le nombre d'étiquettes distinctes. Si une étiquette se répète, elle affecte la même valeur que celle attribuée précédemment.

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html

Si nous devons transmettre ces données au modèle, nous devons encoder la colonne Country dans sa représentation numérique à l'aide de Label Encoder. Après avoir appliqué Label Encoder, nous obtiendrons un résultat comme indiqué ci-dessous

```python
# charger les packages
import pandas as pd
import numpy as np
from sklearn import preprocessing #label encoder
```

```python
#lire le dataset
df = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
```

```python
# Afficher les 5 lignes
df.head()
```

```python
df.describe()
```

```python
#identifier les valeurs uniques
df['variety'].unique()
```

```python
# l'objet label_encoder sait comment comprendre les étiquettes de mots.
label_encoder = preprocessing.LabelEncoder()

# Encoder les étiquettes dans la colonne 'variété'.
df['variety']= label_encoder.fit_transform(df['variety'])

df['variety'].unique()
```

```python
#lire le dataset
iris = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
```

```python
# Afficher les 5 lignes
iris.head()
```

```python
#https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html
#Convertir la variable catégorielle en variables fictives/indicateurs.
dummies = pd.get_dummies(iris.variety)
dummies
```

```python
#Concaténer les variables factices nouvellement créées avec les données chargées et les stocker dans merged_data
merged_data = pd.concat([iris,dummies], axis=1)
merged_data.head()
```

En résumé :

- Utilisez LabelEncoder pour des variables catégorielles ordinales lorsque l'ordre entre les catégories a un sens.
- Utilisez OneHotEncoder pour des variables catégorielles nominales lorsque l'ordre n'a pas de sens et que vous voulez éviter de fausses relations d'ordre.

N'oubliez pas que dans de nombreux cas, vous pourriez avoir besoin de combiner ces techniques avec d'autres méthodes de prétraitement de données, telles que la normalisation ou la standardisation, en fonction des exigences spécifiques de votre problème et des algorithmes que vous utilisez
